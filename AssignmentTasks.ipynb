Create a strong code for password authentication using python. 

```python
password = input("Enter your password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

special_chars = "@#$%&!"

if len(password) < 8:
    print(" Weak password: Must be at least 8 characters long")
else:
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        print("Password is strong. Access granted.")
    else:
        print("Weak password.")
        print("Password must contain:")
        print("- At least one uppercase letter")
        print("- At least one lowercase letter")
        print("- At least one digit")
        print("- At least one special character (@ # $ % & !)")
```

Assignment Name : Logic Builder

Description : Print numbers 1–50 with Fizz/Buzz logic and count occurrences using loops and functions.

```python
def fizz_buzz():
    fizz_count = 0
    buzz_count = 0
    fizzbuzz_count = 0

    for num in range(1, 51):
        if num % 3 == 0 and num % 5 == 0:
            print("FizzBuzz")
            fizzbuzz_count += 1
        elif num % 3 == 0:
            print("Fizz")
            fizz_count += 1
        elif num % 5 == 0:
            print("Buzz")
            buzz_count += 1
        else:
            print(num)

    print("\n--- Counts ---")
    print("Fizz:", fizz_count)
    print("Buzz:", buzz_count)
    print("FizzBuzz:", fizzbuzz_count)


# Function Call
fizz_buzz()
```

*Assignment (19/02/2026)*

*Assignment Name* : Student Data Manager

*Description* : Store data for 5 students using dictionaries, print topper, class average, and assign grades.

```python

students = {
    "Rahul": 85,
    "Sneha": 92,
    "Arjun": 76,
    "Meena": 88,
    "Kiran": 95
}

# Print all student data
print("Student Marks:")
for name, marks in students.items():
    print(name, ":", marks)

# Find Topper
topper = max(students, key=students.get)
print("\nTopper:", topper, "-", students[topper])

# Calculate Class Average
total_marks = sum(students.values())
average = total_marks / len(students)
print("Class Average:", average)

# Assign Grades
print("\nGrades:")
for name, marks in students.items():
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    else:
        grade = "D"
    
    print(name, ":", grade)
```

*Assignment (26/02/2026)*

*Assignment Name* : Data Doctor

*Description* : Clean a dataset by handling missing values, removing duplicates, standardizing text, and explain why cleaning matters.

-------------------------------------------------------

Why Data Cleaning Matters
1. Improves Accuracy

Dirty data leads to wrong predictions.

2. Prevents Bias

Duplicates and errors can mislead models.

3. Ensures Consistency

Standardized text avoids category confusion.

4. Required for ML Models

Most algorithms cannot handle missing or inconsistent data.

5. Better Insights

Clean data = reliable analysis.

```python
import pandas as pd

#  Create Dirty Dataset
data = {
    "Name": ["Rahul", "Sneha", "rahul", "Arjun", None],
    "Age": [20, 21, 20, None, 22],
    "City": ["Hyderabad", "HYDERABAD", "hyderabad", "Chennai", "Chennai"],
    "Marks": [85, 90, 85, 88, None]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Handle Missing Values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Name"].fillna("Unknown", inplace=True)

# Standardize Text (Make lowercase and remove spaces)
df["Name"] = df["Name"].str.lower()
df["City"] = df["City"].str.lower()

# Remove Duplicates
df = df.drop_duplicates()

print("\nCleaned Dataset:")
print(df)
```

*Assignment (28/02/2026)*

*Assignment Name* : Storytelling with Graphs

*Description* : Create bar chart, pie chart, histogram and write a short data story explaining trends..

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 170, 210, 250]

# Bar Chart
plt.figure()
plt.bar(months, sales)
plt.title("Monthly Sales - Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

#Pie Chart
plt.figure()
plt.pie(sales, labels=months, autopct='%1.1f%%')
plt.title("Sales Distribution - Pie Chart")
plt.show()

# Histogram
plt.figure()
plt.hist(sales, bins=5)
plt.title("Sales Distribution - Histogram")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")
plt.show()
```
